```python
import numpy as np
import matplotlib.pyplot as plt

control_freq = 1000                         # frequency in Hz
target_freq = 30                            # frequency in Hz

max_time = 5                                # maximum simulation time
L = 1                                       # length of joint

target_points = max_time * target_freq + 1
target_time = np.linspace(0, max_time, target_points)
target_x = np.full(target_points, 2*L)
target_y = L * np.sin(2 * np.pi * 5 * target_time)

figure, (axs1, axs2) = plt.subplots(2, 1, sharex=True, figsize = (15, 5))
figure.suptitle("Target Trajectory at 30 Hz")
y1_ticks = [1.75, 2, 2.25]
y2_ticks = [-1, -0.5, 0, 0.5, 1]

axs1.plot(target_time, target_x)
axs1.set_yticks(y1_ticks)
y1_tick_labels = [f"{tick/L :.2f}L" for tick in y1_ticks]
axs1.set_yticklabels(y1_tick_labels)
axs1.set_xlabel("time /s")
axs1.set_ylabel("target x position")
axs1.grid(True)

axs2.plot(target_time, target_y, "r")
axs2.set_yticks(y2_ticks)
y2_tick_labels = [f"{tick/L :.2f}L" for tick in y2_ticks]
axs2.set_yticklabels(y2_tick_labels)
axs2.set_xlabel("time /s")
axs2.set_ylabel("target y position")
axs2.grid(True)

plt.show()
```
